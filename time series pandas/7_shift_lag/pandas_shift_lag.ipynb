```python
df['Prev Day Price'] = df['Price'].shift(1)
df
```

```python
df['Price Change'] = df['Price'] - df['Prev Day Price']
df
```

```python
df['5 day return'] =  (df['Price'] - df['Price'].shift(4))*100/df['Price'].shift(4)
df
```

```python
df = df[['Price']]
df
```
